# Data Preprocessing & Feature Engineering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, StandardScaler , MinMaxScaler
```

#Data Exploration & Pre-Processing

```python
# define columns from dataframe
columns = ["age","workclass","fnlwgt","education","education_num","marital_status","occupation","relationship","race","sex","capital_gain","capital_loss","hours_per_week","native_country","income"]
```

```python
df = pd.read_csv("/content/adult_with_headers (1).csv")
print("basic summary statistics")
display(df.describe())
print("\n  Data Tyes",df.dtypes)

```

```python
# Handling the missing values
df.replace("?",np.nan,inplace=True)
```

```python
print("\n Misiing values before imputation",df.isnull().sum())
```

```python
"""no need to impute values as their are no null values in any column"""
```

```python
# Separating columns into numerical(num) & categorical(cat)
num_cols = df.select_dtypes(include = ["int64","float64"]).columns
cat_cols = df.select_dtypes(include = ["object"]).columns
```

```python
# scaling numerical features
df_standard_scaled = df.copy()
df_minmax_scaled = df.copy()

std_scaler = StandardScaler()
df_standard_scaled[num_cols] = std_scaler.fit_transform(df[num_cols])

mm_scaler = MinMaxScaler()
df_minmax_scaled[num_cols] = mm_scaler.fit_transform(df[num_cols])
print("\n scaling completed")
```

# Encoding Techniques

```python
#determining the category counts to decide encoding strategy
print("\n--- Unique Category Counts ---")
# Re-evaluate cat_cols after one-hot encoding in previous steps
cat_cols = df.select_dtypes(include = ["object"]).columns

for col in cat_cols:
    print(f"{col}: {df[col].nunique ()}categories")
```

```python
# apply one-hot encoding(< 5 categories)
less_than_5_cats = [col for col in cat_cols if df[col].nunique() <= 5]
df = pd.get_dummies(df,columns=less_than_5_cats, drop_first = True)
```

```python
#apply label encoding(>=5 categories)
# Re-evaluate cat_cols after one-hot encoding
cat_cols = df.select_dtypes(include = ["object"]).columns
more_than_5_cats = [col for col in cat_cols if df[col].nunique() >= 5]
le= LabelEncoder()
for col in more_than_5_cats:
    df[col] = le.fit_transform(df[col])
print("\n--- Encoding Complete. New Shape:---",df.shape)
```

# Feature Engineering

```python
# New Feature 1 : Net capital
df['net_capital'] = df['capital_gain'] - df['capital_loss']
```

```python
df['net_capital'].head()
```

```python
#New Feature 2: is US native
temp_df = pd.read_csv("/content/adult_with_headers (1).csv")
df["is_US_native"] = temp_df["native_country"].apply(lambda x: 1 if x == "United-States" else 0)
```
