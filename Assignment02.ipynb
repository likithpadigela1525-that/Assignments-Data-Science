### Q11

```python
# Create DataFrame with 10 rows
data = {
    "Name": ["Likith", "Anita", "Kiran", "Sneha", "Arjun", "Priya", "Vikas", "Neha", "Rohit", "Anita"],
    "Age": [21, 22, 23, 21, 24, 22, 23, 21, 25, 22],
    "City": ["Hyderabad", "Mumbai", "Delhi", "Chennai", "Pune", "Kolkata", "Delhi", "Mumbai", "Pune", "Mumbai"],
    "Salary": [50000, 60000, np.nan, 55000, 70000, np.nan, 65000, 62000, 72000, 60000]
}
df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

# Info and Describe
print("\nInfo:")
df.info()

print("\nDescribe:")
print(df.describe())

# Select 'Name' and 'City'
selected = df[['Name', 'City']]
print("\nSelected Columns (Name & City):\n", selected)

# Drop 'City' column
df_dropped = df.drop(columns=['City'])
print("\nAfter Dropping 'City':\n", df_dropped)

# Fill missing Salary with mean
mean_salary = df['Salary'].mean()
df['Salary'].fillna(mean_salary, inplace=True)
print("\nAfter Filling Missing Salary:\n", df)

# Remove duplicate rows
df_no_duplicates = df.drop_duplicates()
print("\nAfter Removing Duplicates:\n", df_no_duplicates)
```
