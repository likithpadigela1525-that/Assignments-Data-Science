```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_csv("Billing_Data.csv")
```

```python
df1
```

```python
df2 = pd.read_csv("Patient_Data.csv")
df2
```

```python
# 1 load patient datset and info()
print("billing info:")
print(df1.info())
print("patient info:")
print(df2.info())
```

```python
df2.describe()
```

```python
# 2 selecting columns for billing ["patient id","department","doctor","bill amount"]
df_selected = df2[["PatientID","Department","Doctor","BillAmount"]]
df_selected
```

```python
# 3 Drop columns "receptionistid","checkin time"
df_clean = df2.drop(columns=["ReceptionistID","CheckinTime"],errors="ignore")
df_clean
```

```python
# 4 total bill amount per department
# 6 adding mean bill to all empty bills
mean_bill = df_clean["BillAmount"].mean()
df_clean["BillAmount"] = df_clean["BillAmount"].fillna(mean_bill)
df_clean
```

```python
#5 remove duplicate patients details
df_clean = df_clean.drop_duplicates(subset="PatientID")
df_clean
```

```python
# 4 groupby total per department
dept_total = df_clean.groupby("Department")["BillAmount"].sum()
dept_total
```

```python
# 7 merge billing dataset with patient dataset on "PatientID"
df_merged = pd.merge(df_clean, df1, on="PatientID", how="inner")
df_merged
```

```python
# 8 add new weekly patient row wise
new_patients = pd.DataFrame({
    'PatientID': [106, 107],
    'Name': ['Rahul', 'Sneha'],
    'Department': ['Cardiology', 'Neurology'],
    'Doctor': ['Dr. Smith', 'Dr. John'],
    'BillAmount': [5500, 4800]
})
df_final = pd.concat([df_merged, new_patients], axis=0, ignore_index=True)
new_patients
```

```python
# add new billing columns columns wise
new_billing_cols = pd.DataFrame({
    'InsuranceCovered': [1200, 1800],
    'FinalAmount': [4300, 3000]
})
new_billing_cols
```

```python
new_billing_cols = new_billing_cols.reindex(df_final.index)

df_final = pd.concat([df_final, new_billing_cols], axis=1)
df_final
```

```python
print("Final Cleaned Dataset:")
print(df_final)
```
